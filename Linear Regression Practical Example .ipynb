### Multicollinearity

1. we can check it through VIF (variance inflation factor)
2. there is no speacial method to check it in sklearn therefore to find it we have to use stats model.


```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['Features']=variables.columns
```

```python
vif.head()
```

### Note:
- VIF = 1 ; no multicollinearity
- 1 < VIF < 5 ; prefectly okay
- 10 < VIF ; very multicollinearity (should be removed)

- since we can see here that year is highly corelated variable it can be removed.

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

# PART 3

### Create Dummy Variables

- for 'n' categoris for a feature we have to create 'n-1' dummies.
- Brands has 7 categories so there will be 6 dummies for Brands
- Engine Type has 4 categorries so there will be 3 dummies for EngineType

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

```python
data_with_dummies.columns
```

```python
## Check the VIF of the features including the dummies
variables_dummies = data_with_dummies[['Mileage', 'EngineV','Brand_BMW', 'Brand_Mercedes-Benz',
       'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota', 'Brand_Volkswagen',
       'Body_hatch', 'Body_other', 'Body_sedan', 'Body_vagon', 'Body_van',
       'Engine Type_Gas', 'Engine Type_Other', 'Engine Type_Petrol',
       'Registration_yes']]
vif_d = pd.DataFrame()
vif_d['VIF'] = [variance_inflation_factor(variables_dummies.values, i) for i in range(variables_dummies.shape[1])]
vif_d['Features']=variables_dummies.columns
```

```python
vif_d
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW', 'Brand_Mercedes-Benz',
       'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota', 'Brand_Volkswagen',
       'Body_hatch', 'Body_other', 'Body_sedan', 'Body_vagon', 'Body_van',
       'Engine Type_Gas', 'Engine Type_Other', 'Engine Type_Petrol',
       'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed
```

# PART 4

### Declare the inputs and the targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

### Scale The Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_text = train_test_split(inputs_scaled, targets, test_size =0.2, random_state = 365)
```

### Create The Regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Tragets(Y-train)', size=15)
plt.ylabel('Predictions(Y-Hat)', size=15)

plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title('Residuals PDF', size=18)
```

```python
reg.score(x_train, y_train)
```

### Finding Weights And Bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns= ['Features'])
reg_summary['Weights']=reg.coef_
reg_summary
```

### Interpretations:

#### as it can be seen that some weights/coefficents are positive while others are negative (continuous Variables)
- A positive Weight Shows that as a feature increases in value, so do the LogPrice and price respectively (EngiveV)
- A negative weight shows that a ]s a feature increases in values, there is a drecrease in the LogPice and the Price (Mileage)

#### the situations for the dummies is quite different (dummy variables)
- since we have droped the categories in each features and created the dummies, therefore the dummies_variables show respective change with respect to the droped variable 
- For Example: in brands the Audi is dropped, so it is taken a 0/reference to other dummies for the Feature Brand (i.e. BMW and Mercedes has values slight increase wrt Audi whereas for other cars there is decrease in values wrt Audi)

#### The size of the Weights
- the bigger the weight the bigger the impact
- For Exxample Mileage seems to be very important as its value is highest and also twice as that of engine value.
- there is no comparision between continuous variable and dummies (dummies are only compared wrt the benchmark dummy)
